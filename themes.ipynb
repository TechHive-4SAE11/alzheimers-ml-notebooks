# 🧠 Thèmes du Projet — Alzheimer & Machine Learning

---

## BO1 — Détection du Risque (Données de Mode de Vie)

> **Dépistage basé sur les habitudes de vie**

| | |
|---|---|
| 🏢 **BO** (Objectif Métier) | **Dépistage Proactif des Risques.** Identifier les individus présentant un risque élevé de développer Alzheimer avant l'apparition des symptômes cliniques, permettant ainsi des interventions précoces sur le mode de vie. |
| 🔬 **DSO** (Objectif Data Science) | Prédire la probabilité (de 0 à 1) qu'un patient soit atteint d'Alzheimer en fonction d'entrées tabulaires comme le régime alimentaire, le sommeil ou l'âge. |
| 🤖 **Type de ML** | Apprentissage Supervisé >>> **Classification Binaire** |

---

## BO2 — Détection de la Sévérité (Scanners IRM)

> **Diagnostic du stade de la maladie à partir d'images**

| | |
|---|---|
| 🏢 **BO** (Objectif Métier) | **Graduation Clinique Automatisée.** Assister les radiologues en classant automatiquement la sévérité de la démence à partir des IRM pour prioriser les cas urgents (Triage). |
| 🔬 **DSO** (Objectif Data Science) | Classer une image d'entrée dans l'une des 4 catégories spécifiques : **Sain**, **Très léger**, **Léger**, **Modéré**. |
| 🤖 **Type de ML** | Apprentissage Supervisé >>>>> **Classification d'Images Multi-classes** |

---

## BO3 — Recommandation (Données de Mode de Vie)

> **Suggestions de changements pour réduire le risque**

| | |
|---|---|
| 🏢 **BO** (Objectif Métier) | **Optimisation Personnalisée de la Santé.** Fournir des conseils concrets et personnalisés aux utilisateurs « à risque » pour modifier leur comportement (ex : *« Augmenter le sommeil d'une heure »*) afin de ressembler à la population « saine ». |
| 🔬 **DSO** (Objectif Data Science) | Segmenter les patients en groupes homogènes à l'aide de techniques de clustering afin d'identifier des profils de risque distincts. Pour chaque cluster, analyser les caractéristiques dominantes et générer des recommandations personnalisées de changements minimaux permettant de rapprocher les individus du profil « sain ». |
| 🤖 **Type de ML** | Apprentissage Non-Supervisé **(Clustering et Recommandation)** |

## Récapitulatif du Déploiement Global

| BO | Service | Artéfacts principaux | Endpoint |
|---|---|---|---|
| **BO1 — Risque clinique** | Random Forest tabulaire + prétraitement sklearn | `models/tabular_model.joblib`, `models/tabular_scaler.joblib`, `models/tabular_pca.joblib`, `models/tabular_features.joblib` | `POST /predict-clinical` |
| **BO2 — Sévérité IRM** | CNN Custom + ResNet50 TensorFlow/Keras | `models/alzheimer_cnn.keras`, `models/alzheimer_resnet.keras` | `POST /predict-mri` |
| **BO3 — Recommandation** | K-Means + NearestNeighbors + table de référence | `alzheimer_api/models/care_clusterer.joblib`, `alzheimer_api/models/care_recommender.joblib`, `alzheimer_api/models/care_scaler.joblib`, `alzheimer_api/models/patient_care_reference.csv` | `POST /recommend-care` |
| **API** | FastAPI + Uvicorn | `alzheimer_api/main.py` | Port 8000 |
| **Interface** | React + Vite | `alzheimer-dashboard/src/App.jsx` | Dashboard web |
| **Conteneurisation** | Docker | `alzheimer_api/Dockerfile` | Déploiement API |
